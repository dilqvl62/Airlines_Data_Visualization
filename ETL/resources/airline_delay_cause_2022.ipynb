# Extract the Airlines_Delay_Cause.csv Data

```python
# Read the data into a pandas DataFrame
airlines_delay_cause_df = pd.read_csv('Airline_Delay_Cause.csv')
airlines_delay_cause_df.head()
```

```python
# Get a brief summary of the airline_delay_cause DataFrame
airlines_delay_cause_df.info()
```

```python
# Get the airline_delay_cause_df columns
airlines_delay_cause_df.columns
```

```python
# get the unique carrie name and airport name 
carrier_name = airlines_delay_cause_df['carrier_name'].unique()
airports = airlines_delay_cause_df['airport_name'].unique()
carrier_name
```

```python
# get the lenght of carrier_name 
print(len(carrier_name))
```

```python
# GET THE LENGHT OF AIRPOTS
print (len(airports))
```

```python
#Rename the columns
airlines_delay_cause_df= airlines_delay_cause_df.rename(columns={'arr_flights': 'total_arriving_flights',\
                                                                 'airport': 'Airport Code',\
                                                                 'carrier_ct':'air_carrier_delay',\
                                                                 'weather_ct': 'wheather_delay', \
                                                                 'nas_ct':'nat_aviation_delay',\
                                                                 'carrier_delay':'carrier_ttl_delay',\
                                                                 'arr_cancelled': 'Cancelled_flights',\
                                                                 'weather_delay': 'weather_ttl_delay',\
                                                                 'nas_delay':'National_aviation_ttl_delay',\
                                                                 'security_delay':'security_ttl_delay',
                                                                 'late_aircraft_delay':"late_aircraft_ttl_delay"
                                                                })

airlines_delay_cause_df.head()
```

```python
#grouping by carrier_name 
total_flights_arrived_df = airlines_delay_cause_df.groupby(['carrier_name'])[['total_arriving_flights', 'air_carrier_delay', 'wheather_delay', 'nat_aviation_delay']].sum()
total_flights_arrived_df
```

```python
#drop off some carrier_names and leave only the top 11 airlines in the USA 
cleaned_top_flights_df = total_flights_arrived_df.drop(['Allegiant Air','Endeavor Air Inc.','Horizon Air','Mesa Airlines Inc.','PSA Airlines Inc.','Republic Airline'])
cleaned_top_flights_df
```

```python
#Exporting the dataframe as a csv file 
cleaned_top_flights_df.to_csv("cleaned_top_flights.csv")
```

```python
#grouping by month and carrier name 
delayed_ttl_duration = airlines_delay_cause_df.groupby(['month', 'carrier_name'])[['carrier_ttl_delay', 'weather_ttl_delay', 'National_aviation_ttl_delay','security_ttl_delay','late_aircraft_ttl_delay']].sum()
delayed_ttl_duration
```

```python
#dropoing all the rows that contain certain carrier_name and leave only the top 11 carrier_name
unwanted_carrier_names = ['Allegiant Air','Endeavor Air Inc.','Horizon Air','Mesa Airlines Inc.','PSA Airlines Inc.','Republic Airline']
delayed_ttl_df = delayed_ttl_duration[~delayed_ttl_duration.index.get_level_values('carrier_name').isin(unwanted_carrier_names)]
delayed_ttl_df
```

```python
# Exporting the dataframe as a csv file
delayed_ttl_df.to_csv('total_time_delay.csv')
```

```python
# creating a state column and a airport_name from the airport_name column 
airlines_delay_cause_df[['states', 'Airport_name']] = airlines_delay_cause_df['airport_name'].str.split(":", expand = True)
airlines_delay_cause_df.tail()
```

```python
# read the data into a pandas dataframe
airport_code_df = pd.read_excel('airports-code@public.xlsx')
airport_code_df.tail()
```

```python
# print the number of aeroport in the dataframe
print(len(airport_code_df['Airport Code'].unique()))
```

```python
# Get the columns 'Airport Code','Airport Name','City Name','Latitude','Longitude'
airport_lat_long = airport_code_df[['Airport Code','Airport Name','City Name','Latitude','Longitude']]
airport_lat_long.tail()
```

```python
#making a copy of the airlines_delay_cause_df
airlines_canceled_df = airlines_delay_cause_df.copy()
airlines_canceled_df.head()
```

```python
# dropping the "airport_name" column 
airlines_canceled_df.drop('airport_name',axis=1, inplace=True)
airlines_canceled_df.head()
```

```python
# Get the colomns 'month', 'carrier_name','Airport Code','Cancelled_flights','states', 'Airport_name'
airlines_canceled_final_df = airlines_canceled_df[['month', 'carrier_name','Airport Code','Cancelled_flights','states', 'Airport_name']]
airlines_canceled_final_df.tail()
```

```python
# Merge the airlines_canceled_final_df with airport_lat_long on "Airport Code"
airports_airlines_df = airlines_canceled_final_df.merge(airport_lat_long, on = 'Airport Code', how= 'left' )
airports_airlines_df.tail()
```

```python
# Droping the "Airport Name" and "Ciry Name"
airports_airlines_df.drop(columns=['Airport Name','City Name'], inplace=True)
```

```python
airports_airlines_df.tail()
```

```python
# Checking for Na values 
airports_airlines_df.isna().sum()
```

```python
# Droping all the Na values 
airports_airlines_df.dropna(inplace=True)
```

```python
#Rename the Airport code to Airport_Code
airports_airlines_df.rename(columns={'Airport Code': 'Airport_Code'}, inplace=True)
```

```python
airports_airlines_df.head()
```

```python
# deleting unwanted carrier names to leave only the top 11 carrier name
unwanted_carrier_names = ['Allegiant Air','Endeavor Air Inc.','Horizon Air','Mesa Airlines Inc.','PSA Airlines Inc.','Republic Airline']
airports_airlines_df = airports_airlines_df[~airports_airlines_df['carrier_name'].isin(unwanted_carrier_names)]
airports_airlines_df.head()
```

```python
# Check for the unique values in carrier_name
airports_airlines_df['carrier_name'].unique()
```

```python
# grouping by Airport_name and summing the canceled
cleaned_airport_airlines_df = airports_airlines_df.groupby(['Airport_name']).agg({  'Airport_Code':'first',\
                                                                                   'Latitude': 'first',\
                                                                                   'Longitude': 'first',\
                                                                                   'Cancelled_flights': 'sum'
                                                                               
                                                                              })
cleaned_airport_airlines_df
```
